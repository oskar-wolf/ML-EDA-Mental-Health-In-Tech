```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df_encoded = pd.read_excel('to_encode.xlsx')
```

```python
df_encoded['is_tech_company'].replace(0.5,2, inplace = True)
```

```python
df_encoded['provides_mental_health_benefits'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['has_discussed_mental_health'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['offers_mental_health_resources'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['is_anonymity_protected'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['ease_of_leave_for_mental_health'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['mental_health_discussion_consequences'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['physical_health_discussion_consequences'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['comfort_discussing_mental_health_with_coworkers'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['comfort_discussing_mental_health_with_supervisor'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['employer_values_mental_equal_physical'].replace(np.nan, 'Self-Employed', inplace= True)
df_encoded['observed_neg_consequences_for_coworkers'].replace(np.nan, 'Self-Employed', inplace= True)
```

```python
#Encode Oridinal Features of Employer Related Information

le = LabelEncoder()

columns_to_encode = [
    'provides_mental_health_benefits', 'has_discussed_mental_health',
    'offers_mental_health_resources', 'ease_of_leave_for_mental_health',
    'is_anonymity_protected', 'mental_health_discussion_consequences',
    'physical_health_discussion_consequences',
    'comfort_discussing_mental_health_with_coworkers',
    'comfort_discussing_mental_health_with_supervisor',
    'employer_values_mental_equal_physical', 'observed_neg_consequences_for_coworkers'
]

for column in columns_to_encode:
    df_encoded[column] = le.fit_transform(df_encoded[column])
```

```python
#Drop Features Of Self-Employed Related Info as it is mosstly nan and insights can be found with EDA on Self-Employed Subset.
columns_to_drop = [
    'medical_coverage_mental_health', 'knows_mental_health_resources',
    'reveals_mental_health_to_clients', 'impact_of_revealing_to_clients',
    'reveals_mental_health_to_coworkers', 'impact_of_revealing_to_coworkers',
    'mental_health_impacts_productivity', 'percentage_work_time_affected_by_mental_health'
]

df_encoded.drop(columns=columns_to_drop,inplace=True)
```
