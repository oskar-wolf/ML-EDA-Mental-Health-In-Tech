```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_self_employed = pd.read_excel(r'C:\Users\mrosk\OneDrive\Desktop\Mental-Health-in-Tech\Subsets\self_employed_subset.xlsx')
```

```python
df_self_employed.drop(df_self_employed.columns[0], axis=1, inplace=True)
df_self_employed
```

```python
pie_chart_variables = [
    'medical_coverage_mental_health',
    'knows_mental_health_resources',
    'reveals_mental_health_to_clients',
    'impact_of_revealing_to_clients',
    'reveals_mental_health_to_coworkers',
    'impact_of_revealing_to_coworkers',
    'mental_health_impacts_productivity',
    'currently_has_mental_health_disorder'
]

def create_pie_chart (data, columns):
    fig, axs = plt.subplots(nrows=len(columns), figsize=(10,5*len(columns)))
    for i, column in enumerate(columns):
        counts=data[column].value_counts()
        axs[i].pie(counts,labels=counts.index,autopct='%1.1f%%',startangle = 90, colors=plt.cm.Paired(np.linspace(0,1,len(counts))))
        axs[i].set_title(column.replace('_', ' ').title(), loc='left', style='italic', fontweight='bold', fontsize=12)
        axs[i].axis('equal')
    plt.tight_layout()
    plt.show()

create_pie_chart(df_self_employed,pie_chart_variables)

plt.figure(figsize=(12,6))
ax = sns.countplot(data = df_self_employed,x='percentage_work_time_affected_by_mental_health', palette='viridis')
ax.set_title('Percentage of Work Time Affected by Mental Health')
ax.set_xlabel('Percentage')
ax.set_ylabel('')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
