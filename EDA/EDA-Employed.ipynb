```python
pie_chart_variables = [
    'provides_mental_health_benefits',
    'has_discussed_mental_health',
    'offers_mental_health_resources',
    'is_anonymity_protected',
    'ease_of_leave_for_mental_health',
    'mental_health_discussion_consequences',
    'physical_health_discussion_consequences',
    'comfort_discussing_mental_health_with_coworkers',
    'comfort_discussing_mental_health_with_supervisor',
    'employer_values_mental_equal_physical',
    'observed_neg_consequences_for_coworkers'
]


def create_pie_charts(data,columns):
    fig, axs = plt.subplots(nrows=len(columns), ncols=1,figsize=(10,5*len(columns)))
    for i, column in enumerate(columns):
        counts = data[column].value_counts()
        axs[i].pie(counts,labels=counts.index,autopct='%1.1f%%',startangle = 90, colors=plt.cm.Paired(np.linspace(0,1,len(counts))))
        axs[i].set_title(column.replace('_',' ').title(), loc='left', style='italic', fontweight='bold', fontsize=12 )
        axs[i].axis('equal')
    plt.tight_layout()
    plt.show()

create_pie_charts(df_employed,pie_chart_variables)
```
